```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1=pd.read_csv("StudentsPerformance.csv")
df1
```

```python
df1["gender"]=df1["gender"].replace('female','F')
```

```python
df1["gender"]=df1["gender"].replace('male','M')
```

```python
df1.isnull().sum()
```

```python
g=df1.groupby("gender")
```

```python
for gender,gd in g:
  print(gender)
  print(gd)
```

# For Female

```python
female=g.get_group("F")
```

```python
female
```

```python
sns.countplot(data=df1,x="parental level of education",hue="race/ethnicity")
plt.xticks(rotation=90)
plt.show()
```

# 🔍 Key Observations:
Group A (orange) consistently has the highest number of students across all parental education levels, especially for:

Associate's degree (nearly 80)

Some college and high school (above 60)

Group D (red) and Group B (blue) have moderate representation across most categories.

Group C (green) has the lowest counts across all education levels, suggesting lower overall representation.

Master's degree has the fewest parents, indicating fewer students come from homes with a master's-educated parent, especially for Group C and E.

Some college and associate's degree are the most common parental education levels across all groups.



# Final Conclusion
The distribution of students by parental education shows that group A dominates in participation, regardless of education level. Higher parental education levels (like master's degree) are less common overall. This could imply socioeconomic factors affecting education levels within different ethnic groups.

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))  # Wider figure for readability

# Plot 1: Math Score
sns.barplot(data=female, x="race/ethnicity", y="math score",
            hue="parental level of education", width=0.6, ax=axes[0])
axes[0].set_title("Math Score by Race/Ethnicity")

# Plot 2: Reading Score
sns.barplot(data=female, x="race/ethnicity", y="reading score",
            hue="parental level of education", width=0.6, ax=axes[1])
axes[1].set_title("Reading Score by Race/Ethnicity")

# Plot 3: Writing Score
sns.barplot(data=female, x="race/ethnicity", y="writing score",
            hue="parental level of education", width=0.6, ax=axes[2])
axes[2].set_title("Writing Score by Race/Ethnicity")

# Adjust layout
plt.tight_layout()
plt.show()
```

# Analysis of the Charts:
The provided charts show Math, Reading, and Writing scores by race/ethnicity and parental level of education. Each group (A to E) likely represents different racial/ethnic groups. The color-coded bars represent the average scores for students based on their parental education level.

Key Observations:
Parental Education Correlation:

In all three subject areas, students whose parents have higher education levels (e.g., bachelor's, master's) generally score higher.

This trend is consistent across all racial/ethnic groups.

Group-Wise Trends:

Group E consistently has higher average scores in all subjects across nearly all education levels.

Group A tends to have the lowest scores, particularly in math and reading.

Group B also performs well, particularly in reading and writing.

Subject-Wise Performance:

Reading and writing scores are generally higher than math scores.

The gap between the highest and lowest performers is most prominent in math.

What This Suggests About Female Performance:
Although the chart does not explicitly show gender (female/male), here are inferences based on educational performance trends (assuming this follows general research patterns):

Female students tend to outperform males in reading and writing, based on large-scale educational studies (e.g., NAEP, PISA).

The consistently higher scores in reading and writing across groups could imply strong female performance, especially if these subjects were dominated by female test-takers in the dataset.

If this dataset reflects typical trends, then:

Female students likely contributed significantly to the higher reading and writing averages seen across all groups.

Conclusion:
Parental education has a strong positive impact on student performance, regardless of race/ethnicity.

Group E performs best overall, while Group A struggles the most.

Reading and writing scores exceed math scores, possibly reflecting the generally better performance of female students in those subjects.

Improving parental education and targeting support for underperforming groups (like Group A) could help reduce achievement gaps.
