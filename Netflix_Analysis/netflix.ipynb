```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as pg
from datetime import datetime
```

```python
dataset=pd.read_csv("Netflix (1).csv")
```

```python
dataset.head()
```

```python
dataset["Date"]=pd.to_datetime(dataset["Date"])
dataset=dataset.set_index("Date")
dataset.head()
```

```python
sns.lineplot(x=dataset.index,y=dataset["Volume"])
plt.title("Trends Between Volume And Year")
# plt.show()
```

```python
dataset.plot(y=['High','Close','Close'],title="Netflix Stock Price")
```

```python
dataset.plot(y='High',title="Netflix Stock Price")
```

```python
dataset.plot(y='Close',title="Netflix Stock Price")
```

```python
dataset.plot(y='Open',title="Netflix Stock Price")
```

# Analyse Month Day And Year Wise 

```python
fig,(x,z,a)=plt.subplots(3,figsize=(15,10))
dataset.groupby(dataset.index.day).mean().plot(y="Volume",ax=x,xlabel="Day")
dataset.groupby(dataset.index.month).mean().plot(y="Volume",ax=z,xlabel="Month")
dataset.groupby(dataset.index.year).mean().plot(y="Volume",ax=a,xlabel="Year")
```

```python
plt.plot(dataset.groupby(dataset.index.day)["Volume"].mean())
plt.xlabel("Day of Month")
plt.ylabel("Average Volume")
plt.show()
```
